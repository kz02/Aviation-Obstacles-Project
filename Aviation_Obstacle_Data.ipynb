## Aviation Obstacle Data Visualization Project
The original data can be located at: https://www.faa.gov/air_traffic/flight_info/aeronav/digital_products/DailyDOF/

The data for this project was accessed 10/29/2025

This project will take data from the FAA website titled Daily Digital Obstacle File (DOF). The DOF tracks obstacles locations and basic descriptions for aviation users. This project will only look into the state of Texas obstacles and display two visualizations to detail the data.

Reference for data detail information: https://www.faa.gov/air_traffic/flight_info/aeronav/obst_data/

```python
#Import pandas
import pandas as pd

#Import csv file and create a dataframe
df_DOF = pd.read_csv('DOF.csv')
#Clean the data by removing NA's
df_clean = df_DOF.dropna()
#Display the Dataframe
#display(df_clean)

# Milestone 3 
#Create dataframe with only Texas data
df_texas = df_clean [df_clean ['STATE'] == 'TX'].reset_index()
#Only keep columns needed, copied into new dataframe
df_tx_clean = df_texas[['OAS', 'CITY', 'LATDEC', 'LONDEC', 'TYPE', 'AGL', 'JDATE']].copy()
#Change the date format to be easily readable
df_tx_clean['DATE']=pd.to_datetime(df_tx_clean['JDATE'], format='%Y%j')
#Drop the prior jdate column
df_tx_clean.drop(columns=['JDATE'], inplace=True)
#Rename columns to be easily readable for later use
df_tx_clean = df_tx_clean.rename(columns={"OAS" : "Obstacle Authoritative Source", "CITY" : "City",
                                          "LATDEC" : "Latitude", "LONDEC" : "Longitude", "TYPE" : "Type", "AGL" : "Above Ground Level", "DATE" : "Date"})
#Display dataframe so far
display(df_tx_clean)
```
